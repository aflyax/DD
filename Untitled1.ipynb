```python
import numpy as np
from numpy.random import rand
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(context='poster', style='dark')

import pandas as pd
```

```python
tot_num = 1000
x = np.arange(0, tot_num/10, 0.1)
y = np.sin(x)

request_arrivals = np.array(sorted(rand(tot_num)*500)+y)
service_needs = rand(tot_num)

requests = np.array(list(zip(request_arrivals, service_needs)))
requests.shape

plt.hist(requests[:,0], range=[1,100], bins=100)

plt.figure()
```
